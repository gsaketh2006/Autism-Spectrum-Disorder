# Autism Prediction on Adults

Data Exploration

```python
# For adult dataset
import arff
import pandas as pd
# Load arff file
with open('Autism-Adult-Data.arff') as f:
    dataset = arff.load(f)
# Convert to DataFrame
adult = pd.DataFrame(dataset['data'], columns=[attr[0] for attr in dataset['attributes']])
# Display 
adult
```

```python
# To know the number of rows and columns in a dataframe
adult.shape
```

```python
# To display all the columns
pd.set_option('display.max_columns',None)
```

```python
# The dataset contains data like ? and I am replacing those values with null by using numpy library
import numpy as np
adult = adult.replace('?', np.nan)
```

```python
# To know how many null values present in the dataset and the datatype of each column
adult.info()
```

```python
# To know what type of data is the dataset is storing
for col in adult.columns:
    print(col,adult[col].unique())
    print("-"*50)
```
